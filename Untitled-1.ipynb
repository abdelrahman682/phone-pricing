```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("phone_data.csv")
df.head()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
for i in df.columns:
    if df[i].dtype == "object":
        df[i] = df[i].str.lower()
        df[i] = df[i].str.replace(" ", "_")
df
```

```python
df = df.drop(['m'], axis=1)
df.info()
```

```python
x = df.drop(['price', 'name', 'brand', 'model', 'touchscreen', 'operating_system', 'wi-fi', 'bluetooth', 'gps', '3g', '4g/_lte'], axis=1)
y = df['price']
my_model = LinearRegression()
my_model.fit(x, y)
my_model.score(x, y)
```
